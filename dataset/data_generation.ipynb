```python
import matplotlib.pyplot as plt
import numpy as np

# 定义时间序列参数
length = 10000  # 时间序列长度
sampling_rate = 100  # 采样率
time = np.linspace(0, length/sampling_rate, length)

# 定义波的数量
num_waves = 5

# 初始化时间序列
time_series = np.zeros_like(time)

# 随机生成波参数并叠加
for _ in range(num_waves):
    amplitude_sine = np.random.uniform(0.5, 2)  # 正弦波随机幅度
    frequency_sine = np.random.uniform(0.1, 5)  # 正弦波随机频率
    phase_sine = np.random.uniform(0, 2 * np.pi)  # 正弦波随机相位
    
    amplitude_cosine = np.random.uniform(0.5, 2)  # 余弦波随机幅度
    frequency_cosine = np.random.uniform(0.1, 5)  # 余弦波随机频率
    phase_cosine = np.random.uniform(0, 2 * np.pi)  # 余弦波随机相位
    
    sine_wave = amplitude_sine * np.sin(2 * np.pi * frequency_sine * time + phase_sine)
    cosine_wave = amplitude_cosine * np.cos(2 * np.pi * frequency_cosine * time + phase_cosine)
    
    time_series += sine_wave + cosine_wave
```

```python

# 可视化结果
plt.plot(time_series[:2000])
plt.title("Randomly Generated Time Series with Independent Sine and Cosine Waves")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```
